```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_of_age = pd.read_csv("AgeGroupDetails.csv")
```

```python
print(data_of_age.head(10))
```

```python
age_on_x = data_of_age.iloc[0:9,1]
```

```python
print(age_on_x)
```

```python
age_on_y = data_of_age.iloc[0:9,2]
```

```python
print(age_on_y)
```

```python
age_group_lst = []
age_case_lst = []
```

```python
for i in range(0,9):
    age_group_lst.append(age_on_x[i])
    age_case_lst.append(age_on_y[i])
```

```python
print(age_group_lst)
```

```python
print(age_case_lst)
```

```python
print(len(age_group_lst) == len(age_case_lst))
```

```python
plt.bar(age_on_x, age_on_y)
plt.ylabel("Age Cases")
plt.xlabel("Age")
plt.title("Covid-19 India bar plot for Age")
plt.show()
```

```python
fig=plt.figure(figsize=(7,7))
plt.pie(age_case_lst,labels=age_group_lst,autopct = '%1.1f%%')
plt.legend()
plt.title("Covid-19 India Pie Chart for Age")
```

```python
data_of_Gender = pd.read_csv("IndividualDetails.csv")
```

```python
print(data_of_Gender.head(5))
```

```python
gender = data_of_Gender.iloc[:,4]
```

```python
print(gender)
```

```python
Total_male=0
Total_female=0
```

```python
for count in range(0,27889):
    if gender[count]=='M':
        Total_male+=1 
    if(gender[count]=='F'):
        Total_female+=1
```

```python
print("Total_male",Total_male)
```

```python
print("Total_female",Total_female)
```

```python
gender_lst = [Total_male, Total_female]
```

```python
print(gender_lst)
```

```python
gender_label=['Male','Female']
```

```python
print(gender_label)
```

```python
plt.pie(gender_lst,labels=gender_label,autopct = '%1.1f%%')
plt.legend()
plt.title('percentage of Gender affected by Covid in India')
```
